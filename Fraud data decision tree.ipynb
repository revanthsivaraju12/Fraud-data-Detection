```python
a["class"]=a["class"].astype("category")
```

```python
b=a.fillna(0)
```

```python
c = pd.get_dummies(b, columns=['browser','source','sex'])
```

```python
c['signup_time'] =  pd.to_datetime(c['signup_time'], format="%m/%d/%y %H:%M")

c['purchase_time'] =  pd.to_datetime(c['purchase_time'], format="%m/%d/%y %H:%M")
```

```python
c['signup_year'] = c['signup_time'].dt.year
c['signup_month'] = c['signup_time'].dt.month
c['signup_month'] = c['signup_time'].dt.dayofweek
c['signup_day'] = c['signup_time'].dt.day
c['signup_hour'] = c['signup_time'].dt.hour
c['signup_minute'] = c['signup_time'].dt.minute
c['purchase_year'] = c['purchase_time'].dt.year
c['purchase_month'] = c['purchase_time'].dt.month
c['purchase_month'] = c['purchase_time'].dt.dayofweek
c['purchase_day'] = c['purchase_time'].dt.day
c['purchase_hour'] = c['purchase_time'].dt.hour
c['purchase_minute'] = c['purchase_time'].dt.minute
```

```python
c['time difference in days']=(c['purchase_time']-c['signup_time']).dt.days
```

```python
df_filtered2 = c.query("device_id=='HGVNRPOHIBZLJ'")
df_filtered2
```

```python
del c['user_id']
del c['device_id']
del c['signup_time']
del c['purchase_time']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=c.iloc[:,c.columns != 'class']
```

```python
y=c['class']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier().fit(X_train,y_train)
```

```python
dt.score(X_train,y_train)
```

```python
#accuracy on test dataset
dt.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python

y_preds = dt.predict(X_test)
```

```python
X_test.head()
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_preds)

#plot roc curve
plt.plot(false_positive_rate,true_positive_rate)
plt.xlabel("False positive Rate")
plt.ylabel("True Positive Rate")

roc_auc = auc(false_positive_rate, true_positive_rate)

roc_auc
```
