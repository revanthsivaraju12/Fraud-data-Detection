```python
a["class"]=a["class"].astype("category")
```

```python
a.isnull()
```

```python
b=a.fillna(0)
```

```python
c = pd.get_dummies(b, columns=['browser','source','sex'])
```

```python
c['signup_time'] =  pd.to_datetime(c['signup_time'], format="%m/%d/%y %H:%M")

c['purchase_time'] =  pd.to_datetime(c['purchase_time'], format="%m/%d/%y %H:%M")
```

```python
c['signup_year'] = c['signup_time'].dt.year
c['signup_month'] = c['signup_time'].dt.month
c['signup_month'] = c['signup_time'].dt.dayofweek
c['signup_day'] = c['signup_time'].dt.day
c['signup_hour'] = c['signup_time'].dt.hour
c['signup_minute'] = c['signup_time'].dt.minute
c['purchase_year'] = c['purchase_time'].dt.year
c['purchase_month'] = c['purchase_time'].dt.month
c['purchase_month'] = c['purchase_time'].dt.dayofweek
c['purchase_day'] = c['purchase_time'].dt.day
c['purchase_hour'] = c['purchase_time'].dt.hour
c['purchase_minute'] = c['purchase_time'].dt.minute
```

```python
c['time difference in days']=(c['purchase_time']-c['signup_time']).dt.days
```

```python
df_filtered2 = c.query("device_id=='HGVNRPOHIBZLJ'")
df_filtered2
```

```python
del c['user_id']
del c['device_id']
del c['signup_time']
del c['purchase_time']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=c.iloc[:,c.columns != 'class']
```

```python
y=c['class']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression(class_weight='balanced')
```

```python
lr=logreg.fit(X_train,y_train)
```

```python
lr.score(X_train,y_train)
```

```python
#make predictions on unseen data
predictions=logreg.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print(classification_report(y_test, predictions,target_names=["Non class", "Class"]))
```

```python
from sklearn.metrics import roc_auc_score,roc_curve,auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, predictions)
roc_auc = auc(false_positive_rate, true_positive_rate)
roc_auc
```

```python
plt.plot(false_positive_rate,true_positive_rate)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show() 
```
